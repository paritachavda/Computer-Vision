```
from matplotlib import pyplot as plt 

histg_1 = cv2.calcHist([img],[0],None,[256],[0,256])
histg_2 = cv2.calcHist([img],[1],None,[256],[0,256])
histg_3 = cv2.calcHist([img],[2],None,[256],[0,256])

plt.plot(histg_1,'r') 
plt.show() 
plt.plot(histg_2,'g') 
plt.show()
plt.plot(histg_3,'b') 
plt.show()
```

```
src = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
dst = cv2.equalizeHist(src)

import numpy as np
res = np.hstack((src, dst))   

cv2_imshow(res)
```

```
histg_1 = cv2.calcHist([src],[0],None,[256],[0,256])
plt.plot(histg_1) 
plt.show()

histg_1 = cv2.calcHist([dst],[0],None,[256],[0,256])
plt.plot(histg_1) 
plt.show()
```

```
def histogram_equalize(img):
    b, g, r = cv2.split(img)
    red = cv2.equalizeHist(r)
    green = cv2.equalizeHist(g)
    blue = cv2.equalizeHist(b)    
    return cv2.merge((blue, green, red)),red,green,blue

img_equ,red,green,blue=histogram_equalize(img)
# cv2_imshow(img_equ)
res = np.hstack((img, img_equ)) 
cv2_imshow(res)

histg_1 = cv2.calcHist([img],[0],None,[256],[0,256])
plt.plot(histg_1) 
plt.show()

histg_1 = cv2.calcHist([img_equ],[0],None,[256],[0,256])
plt.plot(histg_1) 
plt.show()
```
